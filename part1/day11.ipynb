```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
N=30
x=np.arange(N)
y=4*x-7 + (5*np.random.randn(N))
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis],y)
```

```python
yfit=model.predict(x[:,np.newaxis])
```

```python
plt.scatter(x,y,color='#bd1962')
plt.plot(x,yfit,color='#34b5e0');
```

```python
from sklearn.datasets._samples_generator import make_blobs
X,y=make_blobs(n_samples=500,centers=4,cluster_std=1.2)
plt.scatter(X[:,0],X[:,1],c=y,cmap='Dark2');
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
mdl=SVC(gamma='auto')
mdl.fit(X_train,y_train)
```

```python
y_predict=mdl.predict(X_test)
```

```python
acc=100*np.sum(y_predict==y_test)/y_test.size
print(f'Accuracy is : {acc}%')
```

```python
mdl.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mdl=RandomForestClassifier(n_estimators=20)
mdl.fit(X_train,y_train)
```

```python
y_predict=mdl.predict(X_test)
acc=100*np.sum(y_predict==y_test)/y_test.size
print(f'Accuracy is : {acc}%')
```

```python
mdl.score(X_test,y_test)
```
