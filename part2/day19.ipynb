```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig=plt.figure()
ax=plt.axes()
```

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
y=np.sin(x)
ax.plot(x,y,color='r');
```

```python
plt.style.use('seaborn')
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
y3=y1+y2-0.5
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
fig=plt.figure(figsize=(4,3))
ax=plt.axes()
ax.plot(x,y1,color='b',linestyle='solid')
ax.plot(x,y2,color='g',linestyle='dashed')
ax.plot(x,y3,color=(0.3,0.6,0.7),linestyle='dashdot')
plt.show()
```
