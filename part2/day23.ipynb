```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,4))
h1=np.random.normal(0,0.7,1000)
h2=np.random.normal(-5,2,1000)
h3=np.random.normal(5,3,1000)
plt.hist(h1,alpha=0.5,bins=50,histtype='stepfilled')
plt.hist(h2,alpha=0.5,bins=50,histtype='stepfilled')
plt.hist(h3,alpha=0.5,bins=50,histtype='stepfilled');
```

```python
from seaborn import load_dataset
```

```python
iris=load_dataset('iris')
iris.head()
```

```python
plt.figure(figsize=(4,3))
parms=dict(histtype='stepfilled',alpha=0.5,bins=30)
plt.hist(iris.sepal_length,**parms)
plt.hist(iris.sepal_width,**parms)
plt.hist(iris.petal_length,**parms)
plt.hist(iris.petal_width,**parms);
```

```python
fig,ax=plt.subplots(2,3,figsize=(4,4),sharex='col',sharey='row')
for i in range(2):
    for j in range(3):
        ax[i,j].plot(np.random.rand(10),np.random.rand(10),'sr')
        ax[i,j].set(xticks=[],yticks=[])
```

```python
fig,ax=plt.subplots(2,3,figsize=(4,4),sharex='col',sharey='row')
for i in range(2):
    for j in range(3):
        ax[i,j].imshow(np.random.rand(200,200))
        ax[i,j].set(xticks=[],yticks=[])
```

```python
fig,ax=plt.subplots(2,3,figsize=(4,4),sharex='col',sharey='row')
for i in range(2):
    for j in range(3):
        ax[i,j].imshow(np.random.rand(200,200),plt.cm.bone)
        ax[i,j].set(xticks=[],yticks=[])
```

```python
from sklearn.datasets import fetch_lfw_people
```

```python
faces=fetch_lfw_people(min_faces_per_person=30)
```

```python
plt.imshow(faces.images[3]);
```

```python
plt.imshow(faces.images[3],plt.cm.bone);
```

```python
fig,ax=plt.subplots(4,5,figsize=(4,4),sharex='col',sharey='row')
k=0
for i in range(4):
    for j in range(5):
        ax[i,j].imshow(faces.images[k],plt.cm.bone)
        ax[i,j].set(xticks=[],yticks=[])
        k+=1
```
