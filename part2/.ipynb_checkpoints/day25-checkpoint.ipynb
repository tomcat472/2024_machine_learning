```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.DataFrame({'X':np.random.normal(0,2,1000),
                'Y':np.random.normal(3,5,1000),
                'Z':np.random.normal(-4,10,1000)})
df
```

```python
plt.figure(figsize=(4,3))
for col in df.columns:
    plt.hist(df[col],density=True,alpha=0.5,bins=30)
    sns.kdeplot(df[col],shade=True)
```

```python
plt.figure(figsize=(4,4))
for col in df.columns:
    sns.displot(df[col])
```

```python
with sns.axes_style('white'):
    sns.jointplot(x='X',y='Y',data=df,kind='kde')
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
plt.figure(figsize=(4,3))
# sns.relplot(x=iris.sepal_length,y=iris.sepal_width);
sns.relplot(x='sepal_length',y='sepal_width',data=iris);
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',style='species');
```

```python
sns.relplot(x='sepal_width',y='petal_length',
            data=iris,hue='sepal_length',style='species',
           markers=True);
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,kind='line',ci=False,
           hue='species',style='species',dashes=False,markers=True);
```
