```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
markers=['o','.','x','+','v','^','<','>','s','d']
x=np.linspace(-np.pi,np.pi,30)
y1=np.sin(x)
plt.figure(figsize=(4,4),dpi=100)
plt.plot(x,y1,markers[5],color='r',markersize=3.5,linestyle='-.');
```

```python
markers=['o','.','x','+','v','^','<','>','s','d']
plt.figure(figsize=(4,3),dpi=100)
for m in markers:
    plt.plot(np.random.rand(5),np.random.rand(5),m,markersize=4);
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(4,4),dpi=100)
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
sizes=150*np.random.rand(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,cmap=plt.cm.coolwarm)
plt.colorbar();
```

```python
from seaborn import load_dataset
```

```python
iris=load_dataset('iris')
```

```python
iris.head()
```

```python
clrs=np.random.rand(len(iris.species))
clrs[iris.species==iris.species.unique()[0]]=0.2
clrs[iris.species==iris.species.unique()[1]]=0.6
clrs[iris.species==iris.species.unique()[2]]=0.8
plt.figure(figsize=(4,4))
c=np.random.rand(len(iris.species))
plt.scatter(iris.sepal_length,iris.sepal_width,c=clrs,cmap=plt.cm.coolwarm)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width');
```

```python
plt.style.use('seaborn-v0_8-notebook')
plt.figure(figsize=(4,4))
X,Y=np.meshgrid(np.linspace(0,5,50),np.linspace(0,5,40))
Z=2*X-5*np.sin(Y)+2
plt.contour(X,Y,Z,20,cmap='RdGy')
plt.colorbar();
```

```python
plt.figure(figsize=(4,4))
X,Y=np.meshgrid(np.linspace(0,5,50),np.linspace(0,5,40))
Z=2*X-5*np.sin(Y)+2
plt.contourf(X,Y,Z,20,cmap=plt.cm.coolwarm_r)
plt.colorbar();
```

```python
plt.figure(figsize=(4,4))
X,Y=np.meshgrid(np.linspace(5,15,50),np.linspace(5,15,50))
Z=X+(3*Y)*9*X

plt.contourf(X,Y,Z,40,cmap=plt.cm.inferno_r)
plt.colorbar();
```
