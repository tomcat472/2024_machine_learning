```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(6,4))
ax=plt.axes()
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
y3=y1+y2-0.5
ax.plot(x,y1,'-b',linewidth=0.5,label='sin(x)')
ax.plot(x,y2,'--g',linewidth=0.5,label='cos(x)')
ax.plot(x,y3,':r',linewidth=0.5,label='sin(x)+cos(x)-0.5')
plt.legend(loc=3)
plt.axis('equal')
plt.title('sin-cos-curves')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=100)
ax=plt.axes()
ax.plot(x,y1,'--r',label='sin(x)')
ax.plot(x,y2,':b',label='cos(x)')
ax.plot(x,y3,'-k',label='sin(x)+cos(x)-0.5')
ax.set(xlim=(-1,11),ylim=(-5,5),xlabel='X values',ylabel='Y values',
      title='Sin-Cos-Curves')
plt.legend(loc='lower left')
```

```python
plt.figure(figsize=(7,5),dpi=100)
ax=plt.axes()
x=np.linspace(0,20,1500)
y=np.cos(x)
ax.plot(x,y,color='k',linestyle='dashed')
ax.set(xlim=(1,14),ylim=(1.2,-1.2),xlabel='X-values',ylabel='Y-values',title='My Graph')
plt.show()
```

```python
markers=['o','.','x','+','v','^','<','>','s','d']
x=np.linspace(-np.pi,np.pi,30)
y1=np.sin(x)
plt.figure(figsize=(4,4),dpi=100)
plt.plot(x,y1,markers[5],color='r',markersize=3.5);
```
