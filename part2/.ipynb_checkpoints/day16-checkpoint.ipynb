```python
df=pd.DataFrame(np.arange(10).reshape(5,2),columns=['a','b'])
df
```

```python
df.where(df<5,-df)
```

```python
df.where((df<5)|(df%3==3),-df)
```

```python
df.where((df<5)&(df%3==3),-df)
```

```python
idx =(df<5)&(df%3==3)
```

```python
df.where(idx,-df)
```

```python
df[~(df<5)]=-df
```
