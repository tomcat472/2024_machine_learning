```python
dfx=pd.DataFrame({'A':[23,54,34,12,99,0,9,2,1,3],
                 'B':[20,46,7,8,91,45,56,43,11,38]})
dfx
```

```python
dfx=dfx.where((dfx>10),10)
dfx=dfx.where((dfx<30),30)
dfx
```

```python
df1=pd.DataFrame({'E':['B','J','L','S'],
                 'G':['A','E','E','H']})
df2=pd.DataFrame({'E':['L','B','J','S'],
                 'H':[2004,2008,2012,2018]})
df1
```

```python
df2
```

```python
df3=pd.merge(df1,df2)
```

```python
df3
```

```python
df4=pd.DataFrame({'G':['A','E','H'],
                 'S':['C','G','S']})
df4
```

```python
df5=pd.merge(df3,df4,on='G')
```

```python
df5
```

```python
df6=pd.DataFrame({'G':['A','A','E','E','H','H'],
                 'SK':['M','S','C','L','S','O']})
df6
```

```python
pd.merge(df5,df6,on='G')
```
