```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,20,1500)
y=np.cos(x)
plt.plot(x,y,color='orange',linestyle='dashed');
```

```python
fig=plt.figure(figsize=(4,2))
ax=plt.axes()
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
y3=y1+y2-0.5
ax.plot(x,y1,'-b',linewidth=0.5)
ax.plot(x,y2,'--g',linewidth=0.5)
ax.plot(x,y3,':r',linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(4,2),dpi=100)
plt.plot(x,y1,'--r')
plt.plot(x,y2,':b')
plt.plot(x,y3,'_k')
# plt.xlim(-2,12)
# plt.ylim(-1.5,1.5)
plt.axis([-2,11,-2.5,1.5])
# plt.axis('tight')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(4,2),dpi=100)
plt.plot(x,2*x+3,'--k');
plt.xlim(10,0)
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,20,1500)
y=np.cos(x)
plt.xlim(1,14)
plt.ylim(1.2,-1)
plt.plot(x,y,color='orange',linestyle='dashed')
plt.axis('equal');
```
