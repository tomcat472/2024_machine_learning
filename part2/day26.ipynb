```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x="size",y="total_bill",col="smoker",hue="sex",style="sex",data=tips);
```

```python
sns.set(style='ticks',color_codes=True)
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='sex');
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box',hue='sex');
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin',hue='sex');
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box',hue='sex');
```
